```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('tobacco.csv')

# Display the first 5 rows
df.head()
```

```python
tobacco_use_breakdown = df['TopicDesc'].value_counts()

# Plotting the breakdown
plt.figure(figsize=(10, 6))
tobacco_use_breakdown.plot(kind='bar', color='skyblue')
plt.title('Breakdown of Tobacco Use Type Among Youths')
plt.xlabel('Tobacco Use Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('smoker_data.csv')

# Display the first 5 rows
df.head()
```

```python
smoker_chart = df.groupby('City')['Percentage of adults who smoke'].mean().sort_values()

# Plot the bar chart
plt.figure(figsize=(12, 8))
smoker_chart.plot(kind='bar', color='skyblue')
plt.title('Average percentage of smokers by city')
plt.xlabel('city')
plt.ylabel('%')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('state_population_data.csv')

# Setup the figure and subplots
fig, axs = plt.subplots(1, 1, figsize=(12, 18))

# Bar chart for population rates by state
axs.barh(df['State'], df['2022'], color='skyblue')
axs.set_title('Population by State (2022)')
axs.set_xlabel('Population (In Millions)')
axs.set_ylabel('State')

# Adjust layout for better spacing
plt.tight_layout()
plt.show()
```
